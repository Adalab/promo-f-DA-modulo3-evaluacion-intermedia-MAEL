```python
df.columns = df.columns.str.lower()

df
```

```python
def to_float(columna):
    try:
        return float(columna.replace("'","").replace(',',''))
    except:
        return np.nan
```

```python
lista_columnas = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']

for columna in lista_columnas:
    df[columna]=df[columna].apply(to_float)

df.head(40)
```

```python
df.dtypes
```

```python
df['title'] = df['title'].astype(str).str.lower()
df['album'] = df['album'].astype(str).str.lower()
df.head()
```

```python
df['artist'].replace({'$', 's'})
```

```python
album_df = df[df['album_type'] == 'album']
single_df = df[df['album_type'] == 'single']
compilation_df = df[df['album_type'] == 'compilation']
```

```python
grouped = album_df.groupby('artist').agg({'danceability': ['mean', 'std'],
                                           'acousticness': ['mean', 'std'],
                                           'speechiness': ['mean', 'std']})

grouped.columns = ['danceability_mean', 'danceability_std',
                   'acousticness_mean', 'acousticness_std',
                   'speechiness_mean', 'speechiness_std']


danceability_df = grouped[['danceability_mean', 'danceability_std']].sort_values(by='danceability_mean', ascending=False)
acousticness_df = grouped[['acousticness_mean', 'acousticness_std']].sort_values(by='acousticness_mean', ascending=False)
speechiness_df = grouped[['speechiness_mean', 'speechiness_std']].sort_values(by='speechiness_mean', ascending=False)

print("DataFrame de Danceability:")
print(danceability_df)

print("\nDataFrame de Acousticness:")
print(acousticness_df)

print("\nDataFrame de Speechiness:")
print(speechiness_df)
```

```python
grouped = single_df.groupby('artist').agg({'danceability': ['mean', 'std'],
                                           'acousticness': ['mean', 'std'],
                                           'speechiness': ['mean', 'std']})

grouped.columns = ['danceability_mean', 'danceability_std',
                   'acousticness_mean', 'acousticness_std',
                   'speechiness_mean', 'speechiness_std']

danceability_df = grouped[['danceability_mean', 'danceability_std']].sort_values(by='danceability_mean', ascending=False)
acousticness_df = grouped[['acousticness_mean', 'acousticness_std']].sort_values(by='acousticness_mean', ascending=False)
speechiness_df = grouped[['speechiness_mean', 'speechiness_std']].sort_values(by='speechiness_mean', ascending=False)

print("DataFrame de Danceability:")
print(danceability_df)

print("\nDataFrame de Acousticness:")
print(acousticness_df)

print("\nDataFrame de Speechiness:")
print(speechiness_df)
```

```python
grouped = compilation_df.groupby('artist').agg({'danceability': ['mean', 'std'],
                                           'acousticness': ['mean', 'std'],
                                           'speechiness': ['mean', 'std']})

grouped.columns = ['danceability_mean', 'danceability_std',
                   'acousticness_mean', 'acousticness_std',
                   'speechiness_mean', 'speechiness_std']


danceability_df = grouped[['danceability_mean', 'danceability_std']].sort_values(by='danceability_mean', ascending=False)
acousticness_df = grouped[['acousticness_mean', 'acousticness_std']].sort_values(by='acousticness_mean', ascending=False)
speechiness_df = grouped[['speechiness_mean', 'speechiness_std']].sort_values(by='speechiness_mean', ascending=False)

print("DataFrame de Danceability:")
print(danceability_df)

print("\nDataFrame de Acousticness:")
print(acousticness_df)

print("\nDataFrame de Speechiness:")
print(speechiness_df)
```
